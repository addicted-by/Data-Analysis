Импорт необходимых библиотек

```
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats as sts
%matplotlib inline
```

 Выбранное непрерывное распределение - распределение Фишера

```
dfn, dfd = 29, 18 #параметры распределения
fisher = sts.f(dfn, dfd)
x = np.linspace(0,3,100) #отрезок распределения
r = fisher.rvs(size=1000) #выборка, объема 1000
plt.hist(r, density=True, label = 'sample histogram') #построение гистограммы выборки
plt.plot(x, fisher.pdf(x),
       'r-', lw=2, label='f pdf') #построение плотности вероятности
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Для трех значение n (5, 10, 50) генерируется 1000 выборок объема n. Среднее распределения Фишера

image.png
~ 1.125

Дисперсия:

 image.png
 ~ 0.28

```
for n in [5, 10, 50]:
    r_means = []
    for rvn in range(1000):
        r_means.append(fisher.rvs(n).mean())
    plt.hist(r_means, 20, density=True, 
             range=[0, 3], label='means histogram')
    norm_rv = sts.norm(1.125, np.sqrt(0.28/n))
    x = np.linspace(0, 3, 100)
    plt.plot(x, norm_rv.pdf(x),
       'r-', lw=2, label='norm pdf')
    plt.legend()
    plt.title("Sample size %d" % n)
    plt.ylabel('fraction of samples $f(x)$')
    plt.xlabel('sample mean $x$')
    plt.show()
```

```
np.var(fisher.rvs(120000000))
```

При увеличении размера выборки, распределение средних значений начинает сходиться к 1.125, дисперсия при этом падает. Можно сделать вывод, что при увеличении n увеличивается точность аппроксимации.
Нормальное распределение совпадает с гистограммой по ЗБЧ.
